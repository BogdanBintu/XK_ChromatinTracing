```python
def main_analysis2(ifov,return_obj=False,subtract_no=10**8):
    import numpy as np
    import glob,os,sys
    import cv2
    import matplotlib.pyplot as plt
    import tifffile
    sys.path.append(r'C:\Users\Bogdan\Dropbox\ChrisGlass_Project\ChromatinAnalysisV2')
    import StandardChromatinAnalysis_BB as sca
    
    from tqdm.notebook import tqdm
    chr_ = sca.chromatin_analyzer(data_folders = [r'\\132.239.201.25\Raw_data\Glass_MERFISH\CGBB_6_10_2022_KO'],
                          save_folder=r'\\132.239.201.25\Raw_data\Glass_MERFISH\CGBB_6_10_2022_KO_Analysis_v5',
                         H0folder=r'\\132.239.201.25\Raw_data\Glass_MERFISH\CGBB_6_10_2022_KO\H0',extension=r'/data')
    chr_.compute_flat_fields();
    ifovs = np.roll(np.arange(len(chr_.fls)),len(chr_.fls)//3)
    chr_.shape=[60,2900,2900]
    
    #ifov,iQ = ifov_iQ
    
    chr_.set_fov(ifov)
    for iQ in np.arange(len(chr_.Qfolders)):
        chr_.set_hybe_RNA(iQ)
        #def get_Hindex(fld): return int(os.path.basename(fld).split('_')[0][1:])
        #Hindex = get_Hindex(chr_.Rfolders[iQ])
        #subtract = Hindex<subtract_no 
        fit_completed = chr_.check_finished_file()
        if not fit_completed:
            try:
                chr_.compute_cell_segmentation(nstart=1,nend=60,rescz=6,resc=4,sz_min_2d=150,use_gpu=False)
                print("Loading background image")
                chr_.load_background_image()
                print("Loading signal image")
                chr_.load_signal_image()
                chr_.compute_drift()
                chr_.fit_RNA_points_in_cells(th_keep=12,subtract_bk=False)
                chr_.save_fits_RNA(icols=None,plt_val=False)
                chr_.re_th_RNA(icols=[0,1,2],th_s = [1.35]*3,save='_v2')
                chr_.save_fits_RNA(icols=None)
            except Exception as e:
                print("Failed")
                print(e)
    if return_obj:
        return chr_
```
